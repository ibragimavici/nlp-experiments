```python
import spacy
nlp = spacy.load("en_core_web_sm")

def pipeline(text):
    doc = nlp(text)

    lemmas = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return lemmas
```

```python
text1 = "I want to kick some butt!"
text2 = "I always have dreams of a searched schooling of creativity."
text3 = "I am going to kick his butt"
text4 = "There are many benefits with eating butt. Personally I would definitely recommend it!"


nested_texts = [text1, text2, text3, text4]
texts_combined = text1 + text2 + text3 + text4
processed_texts = [" ".join(pipeline(text)) for text in nested_texts]
words_list = []
for w in pipeline(texts_combined):
    if w not in words_list:
        words_list.append(w)
        
        
hmmm_dict = {f"text{i+1}": [0] * len(words_list) for i in range(len(nested_texts))}

text_counter = 0
for text in nested_texts:
    for w in pipeline(text):
        if w in words_list:
            slot = words_list.index(w)
            hmmm_dict[f"text{text_counter+1}"][slot] += 1
    text_counter += 1
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

matrix = list(hmmm_dict.values())
sim = cosine_similarity(matrix)

labels = list(hmmm_dict.keys())
sim_df = pd.DataFrame(sim, index=labels, columns=labels)

print(sim_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(processed_texts)

df = pd.DataFrame(tfidf_matrix.toarray(), 
                  index=[f"text{i+1}" for i in range(len(nested_texts))],
                  columns=vectorizer.get_feature_names_out())

print(df)
```
